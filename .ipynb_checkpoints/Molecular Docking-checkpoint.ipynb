### Get Box from Junaid

### Protein Visualization

```python
view = py3Dmol.view()
view.removeAllModels()
view.setViewStyle({'style':'outline','color':'black','width':0.1})

view.addModel(open('data/4AG8-receptor.pdb','r').read(),format='pdb')
Prot=view.getModel()
Prot.setStyle({'cartoon':{'arrows':True, 'tubes':True, 'style':'oval', 'color':'white'}})
view.addSurface(py3Dmol.VDW,{'opacity':0.6,'color':'white'})

view.addModel(open('data/4AG8-AXI.pdb','r').read(),format='mol2')
ref_m = view.getModel()
ref_m.setStyle({},{'stick':{'colorscheme':'greenCarbon','radius':0.2}})

view.zoomTo()
view.show()
```

### Phrase Docking Result

```python
sdf_files_info = []

for mymol in pybel.readfile("sdf", "data/4AG8_lig_smina_out.sdf"):
    SMINA = mymol.data['minimizedAffinity']
    MW = mymol.exactmass
    ID = mymol.title
    result_df = pd.DataFrame({"Molecule" : ID, 
                             "Molecular Weibht": MW, "SMINA": SMINA}, index = [0])
    sdf_files_info.append(result_df)
    
sdf_files_df = pd.concat(sdf_files_info)
sdf_files_df
    
    
```
